# Test hàng loạt file 

```python
import cv2
from ipykernel import kernelapp as app

for i,image in enumerate(data['images']):
    list_annotation_dict =  [anno_dict for anno_dict in data['annotations'] if anno_dict['image_id'] == image['id']]
    bbox_list = [box['bbox'] for box in list_annotation_dict ]
    img = img=mpimg.imread(path_file + image['file_name'])
    
    #Calculate bbox 
    for bbox in bbox_list:
        x_min, y_min, w, h = bbox
        x_min, x_max, y_min, y_max = abs(int(x_min)), abs(int(x_min + w)), abs(int(y_min)), abs(int(y_min + h))
        cv2.rectangle(img, (x_min, y_min), (x_max, y_max), color=(255, 0, 0), thickness =3)

    cv2.imwrite("/mnt/DeepLearning/hand/hand143_panopticdb/result/"+image['file_name'], img) 
```
