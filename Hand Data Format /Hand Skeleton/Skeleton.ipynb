```python
import os
import os.path
import numpy as np
import json
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (20,20)
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

edges = [[0,1],[1,2],[2,3],[3,4],[0,5],[5,6],[6,7],[7,8],[0,9],[9,10],[10,11],[11,12],[0,13],[13,14],[14,15],[15,16],[0,17],[17,18],[18,19],[19,20]]
```

```python
folderPath ='/media/oggy/DeepLearning/hand/hand143_panopticdb/' #Put your local path here
jsonPath = folderPath +'hands_v143_14817.json'
```

```python
with open(jsonPath, 'r') as file:
    data_all = json.load(file)
    data_all= data_all['root']
```

```python
# Example 
data = data_all[1]
imgPath = folderPath + data['img_paths'];
plt.clf()
im = plt.imread(imgPath)
plt.imshow(im)
```

```python
data
```

```python
# All points of skeleton 
points = data['joint_self']
pts = np.array(data['joint_self'])
points
```

```python
print(points[12])
```

```python
# Point cao nhất và thấp nhất của hand 
plt.plot(points[12][0], points[12][1],'r.')
plt.text(points[12][0], points[12][1], '{0}'.format(12))

plt.plot(points[0][0], points[0][1],'r.')
plt.text(points[0][0], points[0][1], '{0}'.format(0))

plt.imshow(im)
```

```python
for ie, e in enumerate(edges):
    rgb = matplotlib.colors.hsv_to_rgb([ie/float(len(edges)),1.0,1.0])
    plt.plot(pts[e,0],pts[e,1],color=rgb)
plt.imshow(im)
```
