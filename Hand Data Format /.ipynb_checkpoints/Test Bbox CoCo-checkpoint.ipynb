```python
import json 
f = open('person/person_keypoints_train2017.json',) 
data = json.load(f)
```

```python
for ano in data['annotations']:
    for image in data['images']:
        if image['id'] == ano['image_id']:
            file_image = image['file_name']
            id_image = ano['image_id']
            seg_image = ano['segmentation']
            bbox_image = ano['bbox']
            key_image = ano["keypoints"]
    break
    

```
